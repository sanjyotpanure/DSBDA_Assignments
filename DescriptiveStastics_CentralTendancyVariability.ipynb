Mean

```python
#Without Libraries
print("Mean of data without using python libraries: ")
mean=sum(data['Income'])/len(data['Income'])
print(mean)

#With Libraries
print("Mean of data with using libraries: ")
data['Income'].mean()
```

Median

```python
#without libraries
n=len(data["Income"])
if n%2==1:
  median=sorted(data["Income"])[round(0.5*(n+1))]
else:
  x, index = sorted(data["Income"]), round(0.5 * n)
  median = 0.5 * (x[index-1] + x[index])
print("Median without using libraries: ")
print(median)
#with libraries
median=data["Income"].median()
print("Median with using libraries: ")
print(median)
```

Mode

```python
# without libraries
income=data["Income"].values.tolist()
income.sort()
count=dict()
for i in income:
    count[i] = count.get(i, 0) + 1
mode = max(count, key=count.get)
print("Mode without using libraries: ")
print(mode)
#with libraries
mode=data["Income"].mode()[0]
print("Mode with using libraries: ")
mode
```

Variance and Standard Deviation

```python
#without using libraries
n=len(data['Income'])
mean=sum(data['Income'])/n
for i in data['Income']:
  var = sum((item - mean)**2 for item in data['Income']) / (n - 1)
print("Variance value without using libraries: ")
print(var)
print("Variance value without using libraries: ")
std=math.sqrt(var)
print(std)

#with using libraries
print("Variance value with using libraries: ")
print(data['Income'].var())
print("Variance value with using libraries: ")
print(data['Income'].std())
```

Skewness: It is a measure of the symmetry or lack of it of a data sample. The skewness value can be positive, negative, or undefined.

Highly skewed distribution: value<-1 or value>+1

Moderately skewed distribution: -1>value>-1/2 or 1>value>1/2

Approximately symmetric distribution: -1/2>value>1/2

```python
#without libraries
skew= (sum((item - mean)**3 for item in data['Income'])* n / ((n - 1) * (n - 2) * std**3))
print("Skewness without using libraries: ")
print(skew)
#with libraries
skew=data.skew()[0]
print("Skewness with using libraries: ")
print(skew)
```

```python
norm = data
norm.plot(kind = 'density')
print('This distribution has skew', norm.skew())
print('This distribution has kurtosis', norm.kurt())
```

Calculating measures of correlation

covariance and correlation coefficient

Covariance indicates the strength and direction of a relationship between the pairs of variables.

When the covariance is positive, the correlation is Positive. A higher value of covariance indicates stronger relationship.

When the covariance is negative, the correlation is Negative. A lower value of covariance indicates a stronger relationship.

When the covariance is close to zero, it indicates that correlation is weak.

```python
iris.corr()
```

```python
iris.cov()
```

```python
iris.groupby('Species').cov()
```

Correlation Coefficient also known as Pearsons Correlation Coefficient is another measure to find out the relation between two variables.

The value  > 0 indicates positive correlation.

The value  < 0 indicates negative correlation.

The value r = 1 is the minimum possible value of . It corresponds to a perfect negative linear relationship between variables.
